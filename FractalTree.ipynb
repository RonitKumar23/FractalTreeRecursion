# This fractal tree is made by using Recursive function

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
p1=[200,0]
p2=[200,50]
```

```python
plt.plot([p1[0],p2[0]],[p1[1],p2[1]],linewidth=10)
```

```python
def midpoint(x1,x2):
    mp=[0,0]
    mp[0]=(x1[0]+x2[0])/2
    mp[1]=(x1[1]+x2[1])/2
    return mp
q=midpoint(p1,p2)
```

```python
n=0
k=7  #This is for the width or tree 
def FractalTree(n,k,point, radians, origin):

    x, y = point
    ox, oy = origin

    qx = ox + math.cos(radians) * (x - ox) - math.sin(radians) * (y - oy)
    qy = oy + math.sin(radians) * (x - ox) + math.cos(radians) * (y - oy)
    
    qx1 = ox + math.cos(-radians) * (x - ox) - math.sin(-radians) * (y - oy)
    qy1 = oy + math.sin(-radians) * (x - ox) + math.cos(-radians) * (y - oy)
    
    
    plt.plot([origin[0],qx],[origin[1],qy],linewidth=k)
    plt.plot([origin[0],qx1],[origin[1],qy1],linewidth=k)
    plt.plot([p1[0],p2[0]],[p1[1],p2[1]],linewidth=10)
    
    #plt.show()
    qr=[qx,qy]
    qs=[qx1,qy1]
    point=midpoint(origin,qr)
    point1=midpoint(origin,qs)
    
   
    if n<6:
        
        origin1=qx1,qy1
        origin=qx,qy
        #rotate_around_point(n+1,q,math.radians(115),p2)
        FractalTree(n+1,k-1,point,math.radians(135),origin)
        FractalTree(n+1,k-1,point1,math.radians(135),origin1)


FractalTree(n+1,k-1,q,math.radians(115),p2)
```
